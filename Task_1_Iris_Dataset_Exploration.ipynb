# Task 1: Exploring and Visualizing the Iris Dataset

## Objective
The objective of this task is to understand how to load, inspect, summarize, and visualize a dataset using Python.

## Dataset
**Iris Dataset** – A well-known dataset containing measurements of iris flowers belonging to three species:
- Setosa
- Versicolor
- Virginica

## Tools & Libraries Used
- pandas
- seaborn
- matplotlib

## Expected Outcomes
- Load and inspect the dataset
- Understand dataset structure
- Generate summary statistics
- Create scatter plots, histograms, and box plots
- Interpret the visualizations


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

plt.style.use('default')
%matplotlib inline
```

## Step 1: Load the Dataset

The Iris dataset is loaded directly from Seaborn's built-in datasets collection.


```python
iris = sns.load_dataset('iris')
iris.head()
```

## Step 2: Display Dataset Structure

We examine:
- Number of rows and columns
- Column names
- First few records


```python
print("Dataset Shape:")
print(iris.shape)

print("\nColumn Names:")
print(list(iris.columns))

print("\nFirst Five Rows:")
display(iris.head())
```

## Step 3: Data Summary

Basic statistical information helps us understand:
- Mean
- Standard deviation
- Minimum and maximum values
- Quartiles


```python
iris.describe()
```

## Step 4: Check Missing Values

Before analysis, it is important to identify missing values.


```python
iris.isnull().sum()
```

## Visualization 1: Scatter Plot

### Purpose
A scatter plot helps analyze relationships between variables.

### Variables Selected
- Sepal Length
- Sepal Width

### Interpretation
Different species tend to form clusters, indicating distinguishable characteristics.


```python
plt.figure(figsize=(8,5))
sns.scatterplot(
    data=iris,
    x='sepal_length',
    y='sepal_width',
    hue='species'
)
plt.title('Sepal Length vs Sepal Width')
plt.show()
```

## Visualization 2: Histogram

### Purpose
A histogram shows the distribution of a variable.

### Variable Selected
- Petal Length

### Interpretation
The distribution highlights how petal lengths vary across flowers and indicates possible grouping patterns.


```python
plt.figure(figsize=(8,5))
sns.histplot(
    iris['petal_length'],
    bins=15,
    kde=True
)
plt.title('Distribution of Petal Length')
plt.xlabel('Petal Length')
plt.show()
```

## Visualization 3: Box Plot

### Purpose
A box plot helps:
- Detect outliers
- Understand spread of values
- Compare distributions across categories

### Variable Selected
- Petal Length by Species


```python
plt.figure(figsize=(8,5))
sns.boxplot(
    data=iris,
    x='species',
    y='petal_length'
)
plt.title('Petal Length by Species')
plt.show()
```

## Findings and Insights

### Dataset Overview
- The Iris dataset contains 150 observations and 5 variables.
- Three flower species are represented equally.

### Scatter Plot Insights
- Species show visible clustering.
- Sepal dimensions help differentiate flower types.

### Histogram Insights
- Petal length is not uniformly distributed.
- Multiple peaks suggest differences among species.

### Box Plot Insights
- Petal lengths vary significantly between species.
- The box plot clearly highlights spread and potential outliers.

## Conclusion
The Iris dataset was successfully loaded, explored, summarized, and visualized. Basic exploratory data analysis (EDA) techniques were applied using pandas, matplotlib, and seaborn to better understand the structure and characteristics of the data.
